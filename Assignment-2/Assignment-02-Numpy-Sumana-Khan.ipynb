#### Method 2: Using generator with limit

```python
def generate_odd_numbers(limit = None):
    num = 1
    while True:
        if limit is not None and num>limit:
            break
        yield num
        num+=2
```

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
def generate_n_array(n:int):
    """
    Description: Creates an n X n array with each element value as i+j (row and column position)

    Input: 
    - n : a positive integer, the dimention of the reslting array

    Returns:
    - an array with size n x n and each element to be i+j (row and column position)   

    """
    try:
        if not isinstance(n, int) or n<=0:
            raise ValueError("n must be positive integer") 
        
        a = np.arange(n)
        b = np.arange(n).reshape(n,-1)
        z = a+b
        return z
    except ValueError as err:
        print(err)
    except Exception as e:
        print(e)
```

```python
print(generate_n_array(5))
```

### methosd 2 - using loop and empty array

```python

def generate_array(n:int):
    """
    Description: Creates an n X n array with each element value as i+j (row and column position)

    Input: 
    - n : a positive integer, the dimention of the reslting array

    Returns:
    - an array with size n x n and each element to be i+j (row and column position)   

    """
    try:
        if not isinstance(n, int) or n<=0:
            raise ValueError("n must be positive integer") 
        
        x= np.empty((n,n), dtype=int)
        for i in range(n):
            for j in range(n):
                x[i,j] = i+j
        return x
    except ValueError as err:
        print(err)
    except Exception as e:
        print(e)
```

```python

output_array = generate_array(5)
if output_array is not None:
    print(output_array)
    print(f"The resulting array size is: {output_array.shape}")
```
