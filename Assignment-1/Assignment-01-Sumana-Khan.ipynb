#### Question-1

Write a function `second_largest(numbers)` that takes a list of numbers and returns the second largest number in the list. Assume the list has at least two unique numbers.

Example:
```python
numbers = [10, 5, 20, 8, 15]
second = second_largest(numbers)
print(second)
# Expected output: 15
```

````python
def second_largest(num_list):
    unique_num_list = list(set(num_list)) # removes duplicates if any
    
    if len(unique_num_list) >1:
        return sorted(unique_num_list, reverse=True)[1]
    else:
        print(f"There is only one number in the list {num_list}")
        return
````

````python
numbers = [10, 5, 20, 8, 15]
second = second_largest(numbers)
print(f"Second largest number in the list: {second}")

# numbers1 = [67,23,54,54,8,56,34,54,34]
# print(second_largest(numbers1))

# numbers2 =[5]
# print(second_largest(numbers2))

# number3 = [5,1]
# print(second_largest(number3))
````

#### Question-2

Write a function `count_vowels(text)` that takes a string as input and returns the number of vowels (a, e, i, o, u - case-insensitive) present in the string.

Example:
```python
text = "Hello World"
vowel_count = count_vowels(text)
print(vowel_count)
# Expected output: 3
```


````python
def count_vowels(text):
    vowels = ['a','e','i','o','u']
    count=0
    for c in text.lower():
        if c in vowels:
            count+=1
    return count
````

````python
text = "Hello World"
vowel_count = count_vowels(text)
print(f"Number of vowels in the '{text}' text is {vowel_count}")
````

#### Question-3

Write a Python program to print all the prime numbers between 1 and N (inclusive). Where N is a positive integer and an input from user.

Example:

```python
N = 20
# Expected output: 2,3,5,7,11,13,17,19

```

````python
def primes(number):
    all_prior_numbers = [num for num in range(2,number)]
    for all_number in all_prior_numbers:
        if number % all_number == 0:
            return False
    return True
````

````python
N=20
prime_numbers = []
if N > 0:
    for n in range(2, N+1):
        if primes(n):
            prime_numbers.append(n)
print(prime_numbers)
````

Example:
```python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
# Expected output:
# True
# False
```

````python
def is_anagram(str1, str2):
    str1_list = [c for c in str1]
    str2_list = [c for c in str2]
    if sorted(str1_list) == sorted(str2_list):
        #print(f"'{str2}' is anagram of '{str1}'")
        return True
    else:
        return False
````

````python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
````

#### Question-7

Write a function `rotate_list(lst, k)` that rotates a given list lst to the right by `k` positions in-place. Assume `k` is a non-negative integer and can be greater than the length of the list.

Example:

```python
my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
# Expected output: [4, 5, 1, 2, 3]
```

````python
def rotate_list(lst, k):
    assert k>0, "position should be only positive number"

    # print(lst[-k:])
    # print(lst[:-k])
    n = len(lst)
    if k>n:
        k=k%n
    return lst[-k:] + lst[:-k]

````

````python
my_list = [1, 2, 3, 4, 5]
shifted_list = rotate_list(my_list, 2)
print(shifted_list)
````

#### Question-9

Write a Python program that prints the first `n` rows of Pascal's triangle. `n` is a positive integer and an user input. To know about Pascal's triangle see the [wikipedia article](https://en.wikipedia.org/wiki/Pascal%27s_triangle): 

Example:

```python
n = 5
# Expected output:
#     1
#    1 1
#   1 2 1
#  1 3 3 1
# 1 4 6 4 1
```

````python
def pascal_triangle(n):
    triangle_list = []
    for row in range(1, n+1):
        element = []
        for col in range(1, row+1): ## row 4, col=1,2,3,4 # row 5, col =1,2,3,4,5
            if col == 1 or col == row :
                element.append(1)
            else:
                new_col = triangle_list[row-2][col-2]+triangle_list[row-2][col-1]
                element.append(new_col)
                #print(element)
        triangle_list.append(element)
    return triangle_list
````

````python
try:
    n = int(input("Enter number of rows of Pascal's Triangle").strip())
    assert n > 0, "Number should be positive integer"

    triangle_list = pascal_triangle(n)
    spaces = len(triangle_list)
    for lst in triangle_list:
        print(" "*spaces, end=" ")
        for j in range(len(lst)):
            print(lst[j], end=" ")
        spaces -=1  
        print()
except ValueError as error:
    print(f"Error: {error}.\nPlease enter positive integer. ")
except AssertionError as err:
    print(err)
except Exception as e:
    print(e)
````

#### Question-10

Write a function `reverse_the_words(s)` which will take a string `s` as input and reverses the words in the string. 

Example:

```python
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
# Expected output: challenging and awesome is Python
```

````python
def reverse_the_words(text):
    words = text.split(" ")
    reversed_words = words[-1::-1]
    return " ".join(reversed_words)

````

````python
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
````

Example:
```python
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)

# Expected output (output can be in any order): {"this": 2, "city": 2, "is": 2, "a": 1, "beautiful": 1, "where": 1, "i": 1, "live": 1}
```

#### Method 1

````python
from collections import defaultdict
import re

def word_counter(text):
    cleaned_text = re.sub(r'[^\w\s]','',text)
    # #print(cleaned_text)
    words = cleaned_text.split(" ")
    
    word_counter = defaultdict(int)
    for word in words:
        word_counter[word] +=1
    return dict(word_counter)
````
