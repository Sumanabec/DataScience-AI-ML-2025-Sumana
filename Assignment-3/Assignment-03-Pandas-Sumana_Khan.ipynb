#### 7. What are the top 5 most populated Counties according to the 2010 Census?

```python
top_five_populated_counties_2010 = pop.sort_values(by="2010Census", ascending=False).head().reset_index(drop=True)
top_five_populated_counties_2010
```

```python
top_five_populated_counties_2010["County"]
```

#### 8. What are the top 5 most populated States according to the 2010 Census?

```python
top_5_most_populated_state_2010 =  pop.groupby("State").sum().sort_values(by="2010Census", ascending=False).reset_index()[["State","2010Census"]].head()
top_5_most_populated_state_2010
```

#### 9. How many Counties have 2010 populations greater than 1 million?

```python
print(f"Number of Counties with 2010 population greater than 1 million: {pop[pop["2010Census"] > 1000000]["County"].count()}")

```

#### 10. How many Counties don't have the word "County" in their name?

```python
print(f"No of counties don't have the word 'County' in their name: {pop[~pop["County"].str.contains("County")]["County"].count()}")
```

#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
pop["population_pct_change"] = round(((pop["2017PopEstimate"] - pop["2010Census"])/pop["2017PopEstimate"])*100, 2)
pop

```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
state_pop = pop.groupby("State").agg({"2010Census": "sum","2017PopEstimate":"sum"}).reset_index()
```

```python
state_pop["population_pct_change"] = round(((state_pop["2017PopEstimate"] - state_pop["2010Census"])/state_pop["2017PopEstimate"])*100, 2)
```

```python
state_pop.sort_values(by="population_pct_change", ascending=False).head()[["State", "population_pct_change"]].reset_index(drop=True)
```
